```python
def coordinates(r=1.0, a=0.0):
    """computes the coordinate pair
    of a circle of radius r and angle a
    input:
    r: radius as float
    a: angle in radius as float
    output:
    tuple(x,y) as float
    """
    return r * cos(a), r*sin(a)

c = coordinates(a=pi/4.0)
print(c)
```
